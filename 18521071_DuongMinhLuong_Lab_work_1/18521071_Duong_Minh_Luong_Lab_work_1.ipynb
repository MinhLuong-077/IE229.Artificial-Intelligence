# 問題 2 
次の計算を行うプログラムを作成し、答えを求めよ。

$\sum_{x=1}^{1000}(x^3+3x^2+5x+9)$

```python
s = 0

def func(x):
  return x**3 + 3*x**2 + 5*x + 9

for i in range(1,101):
  s = s + func(i)
  
print(s)
```

# 問題 3 
与えられたリストを逆順に して返す関数 reverse を作成せよ。

x = [10, 20, 30, 40, 50]
=> [50, 40, 30, 20, 10]

```python
x = [10, 20, 30, 40, 50]
def reverse(x):
  y = []
  for i in range(1,len(x)):
    y.append(x[-i])
  y.append(x[0])
  return y
reverse(x)
```
